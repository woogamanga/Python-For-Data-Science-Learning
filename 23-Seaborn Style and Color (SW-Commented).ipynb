# Style and Color

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Styles

You can set particular styles:

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

## Spine Removal

```python
sns.countplot(x='sex',data=tips)
sns.despine() #this removes the top spine and right spine
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True) #removes top, right and left
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True) #this removes all spines (top,bottom,left,right)
```

## Size and Aspect

You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
# Non Grid Plot
plt.figure(figsize=(12,3)) #this matplotlib call will be used by seaborn
sns.countplot(x='sex',data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,size=2,aspect=3)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='BrBG')
```

## Scale and Context

The set_context() allows you to override default parameters:

```python
sns.set_context('poster',font_scale=3)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```

```python
sns.set_context('paper',font_scale=3)
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context(context=None)
```
