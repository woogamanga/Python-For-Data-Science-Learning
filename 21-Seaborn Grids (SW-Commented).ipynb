# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

earlier we learnt about pairplot which automates the plotting of pairwise relationships between numerical variables. PairGrid works in a similar fashion but offers more control

```python
sns.pairplot(iris) #usually just scatterplots and histograms 
```

## PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# Just the Grid (empty- kinda just makes subplots for us)
sns.PairGrid(iris)
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g2 = sns.PairGrid(iris)
g2.map_diag(sns.kdeplot)
g2.map_upper(sns.boxplot)
g2.map_lower(plt.scatter)
```

## pairplot

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Just the Grid
g = sns.FacetGrid(tips, col="time", row="smoker") 
#specifying col and row here is similar to specifying nrows and ncols for plt.subplots()
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g2 = sns.FacetGrid(tips, col="time",  row="smoker")
g2 = g2.map(sns.distplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice how the arguments come after plt.scatter call
#plt.scatter for instance, takes two arguments 
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## JointGrid

JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips) #this creates A grid
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #this is an alternative to the JointGrid shown above
```

Reference the documentation as necessary for grid types, but most of the time you'll just use the easier plots discussed earlier.
